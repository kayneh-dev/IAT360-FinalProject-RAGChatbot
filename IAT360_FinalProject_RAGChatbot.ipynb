Load and convert PDFs with Docling

```python
from docling.document_converter import DocumentConverter

converter = DocumentConverter()

raw_docs = []  # [ {id, source_path, markdown}, ... ]

for pdf_path in PDF_DIR.glob("*.pdf"):
    print("Converting:", pdf_path)
    result = converter.convert(str(pdf_path))
    markdown_text = result.document.export_to_markdown()
    raw_docs.append({
        "id": pdf_path.stem,          # e.g., 'sfu_calendar_2025_fall'
        "source_path": str(pdf_path),
        "markdown": markdown_text,
    })

len(raw_docs), [d["id"] for d in raw_docs][:10]
```

Chunk the documents

```python
import re
from typing import List, Dict

def split_by_headings(markdown: str) -> List[str]:
    # Split on markdown headings while keeping them in the text
    parts = re.split(r"(?=^#{1,6}\s)", markdown, flags=re.MULTILINE)
    return [p.strip() for p in parts if p.strip()]

def chunk_text(text: str, max_chars: int = 1000) -> List[str]:
    lines = text.splitlines()
    chunks = []
    current = []
    current_len = 0

    for line in lines:
        if current_len + len(line) + 1 > max_chars:
            if current:
                chunks.append("\n".join(current).strip())
            current = [line]
            current_len = len(line)
        else:
            current.append(line)
            current_len += len(line) + 1

    if current:
        chunks.append("\n".join(current).strip())

    return chunks

# Build the chunks with metadata
chunks: List[Dict] = []

for doc in raw_docs:
    sections = split_by_headings(doc["markdown"])
    for sec in sections:
        for c in chunk_text(sec, max_chars=1000):
            chunks.append({
                "doc_id": doc["id"],
                "text": c,
                "source": doc["id"],  # for later URLs
            })

len(chunks)
```

Create embeddings + FAISS index

```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

# 1. Load embedding model (small & fast)
embed_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# 2. Embed all chunks
corpus_texts = [c["text"] for c in chunks]
corpus_embeddings = embed_model.encode(corpus_texts, convert_to_numpy=True, show_progress_bar=True)

# 3. Normalize embeddings (so inner product = cosine similarity)
corpus_norms = np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
corpus_embeddings = corpus_embeddings / corpus_norms

# 4. Build FAISS index
dim = corpus_embeddings.shape[1]
index = faiss.IndexFlatIP(dim)
index.add(corpus_embeddings)

print("Index size:", index.ntotal)
```

retrieval helper

```python
def retrieve_chunks(query: str, k: int = 5):
    q_emb = embed_model.encode([query], convert_to_numpy=True)
    q_emb = q_emb / np.linalg.norm(q_emb, axis=1, keepdims=True)
    scores, indices = index.search(q_emb, k)

    results = []
    for rank, (idx, score) in enumerate(zip(indices[0], scores[0]), start=1):
        if idx == -1:
            continue
        c = chunks[idx]
        results.append({
            "rank": rank,
            "score": float(score),
            "chunk_index": int(idx),
            "text": c["text"],
            "source": c["source"],
        })
    return results

# Quick test:
test_query = "What is the late course withdrawal deadline for the Spring term?"
for r in retrieve_chunks(test_query, k=3):
    print("RANK", r["rank"], "SOURCE", r["source"], "SCORE", r["score"])
    print(r["text"][:250], "...\n")
    print("-" * 80)
```

Load Llama 2-7B chat model

```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf" # meta-llama/Llama-2-7b-hf

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)
```

```python
from transformers import pipeline

llama_pipeline = pipeline(
    "text-generation",  # LLM task
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

Build the RAG prompt and Llama response function

```python
SYSTEM_PROMPT = """You are an SFU academic assistant.
You ONLY answer questions using the SFU documents provided in CONTEXT.
If the answer is not in the context, say you don’t know and suggest contacting SFU advisors.
Always:
- Answer in plain, student-friendly English.
- Include citations like [1], [2] that refer to the numbered context snippets.
- Do NOT invent deadlines, GPA thresholds, or policy rules.
"""

def build_rag_prompt(question: str, retrieved: list) -> str:
    context_parts = []
    for r in retrieved:
        context_parts.append(
            f"[{r['rank']}] Source: {r['source']}\n{r['text']}"
        )

    context_block = "\n\n---\n\n".join(context_parts)

    prompt = f"""{SYSTEM_PROMPT}

CONTEXT:
{context_block}

QUESTION:
{question}

INSTRUCTIONS:
- Use only the information in CONTEXT.
- Quote key sentences when helpful.
- Add citations [1], [2], ... after the sentences they support.
- If the context is conflicting or unclear, say so.

ANSWER:"""
    return prompt
```

```python
def get_llama_response(question: str, top_k: int = 5):
    retrieved = retrieve_chunks(question, k=top_k)
    if not retrieved:
        print("Chatbot: I couldn't find any relevant SFU information in my documents. Please check the SFU website or contact an advisor.")
        return

    prompt = build_rag_prompt(question, retrieved)

    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_new_tokens=256,
    )

    full_text = sequences[0]["generated_text"]
    # Sometimes the model echos the prompt; simple split to show only after ANSWER:
    if "ANSWER:" in full_text:
        answer = full_text.split("ANSWER:", 1)[1].strip()
    else:
        answer = full_text

    print("Chatbot:", answer)
```

```python
def is_on_topic(question: str) -> bool:
    sfu_keywords = ["sfu", "course", "enrol", "enroll", "co-op", "coop", "calendar", "academic", "faculty"]
    q = question.lower()
    return any(kw in q for kw in sfu_keywords)

print("SFU RAG Chatbot. Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower().strip() in ["bye", "quit", "exit"]:
        print("Chatbot: Goodbye!")
        break

    if not is_on_topic(user_input):
        print("Chatbot: I can only answer questions about SFU academic information.")
        continue

    get_llama_response(user_input)
```
